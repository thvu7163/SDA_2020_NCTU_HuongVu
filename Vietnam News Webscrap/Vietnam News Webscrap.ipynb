```python
import requests
from bs4 import BeautifulSoup
import pandas
import numpy as np
```

```python
import os
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
url= 'https://vietnamnet.vn/vn/kinh-doanh/tai-chinh/'
page = requests.get(url)
site = BeautifulSoup(page.content, 'html.parser')
contents = site.find_all('h3')
titles = [content.find('a').text for content in contents]
text = "".join(titles)
```

```python
mask=np.array(Image.open(path.join(os.getcwd(),"vietnam_shape.jpg")))
wc=WordCloud(mask=mask,mode='RGBA', background_color=None)
wc.generate(text)

plt.figure(figsize=(50,50))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off") 
plt.show()
```
